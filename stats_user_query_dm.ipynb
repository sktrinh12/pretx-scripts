```r
library(ggplot2)
library(dplyr)
install.packages("viridis")
install.packages("svglite")
library(viridis)
df <- read.csv("/home/jovyan/work/Documents/RCH/USER_QUERY_OUTPUT.csv")
```

```r
head(df)
```

```r
df %>% select(ISID, USER_QUERY_NAME, LAST_RUN, PROJECT_FORMS_NAME, PROJECTS_DESCRIPTION)
```

```r
df %>% select(PROJECT_FORMS_NAME) %>% distinct()
```

```r
df %>%
  select(ISID, USER_QUERY_NAME, LAST_RUN, PROJECT_FORMS_NAME, PROJECTS_DESCRIPTION) %>%
  filter(grepl("Gateway", PROJECT_FORMS_NAME, ignore.case = TRUE))
```

```r
options(repr.plot.width = 16, repr.plot.height = 12)
```

```r
from_date = min(df$LAST_RUN, na.rm = TRUE)
to_date = max(df$LAST_RUN, na.rm = TRUE)
```

```r
distinct_user_counts <- df %>%
  group_by(PROJECTS_DESCRIPTION) %>%
  summarise(DistinctUserCount = n_distinct(ISID), .groups = "drop") %>%
  arrange(desc(DistinctUserCount))  # Order by distinct user count

# Create a mapping to order PROJECTS_DESCRIPTION in the heatmap
df$PROJECTS_DESCRIPTION <- factor(df$PROJECTS_DESCRIPTION, 
                                   levels = distinct_user_counts$PROJECTS_DESCRIPTION)
df_summary <- df %>%
  group_by(ISID, PROJECTS_DESCRIPTION) %>%
  summarise(Count = n(), .groups = "drop")  # Summarise and drop grouping

# Creating the heatmap
plt <- ggplot(df_summary, aes(y = ISID, x = PROJECTS_DESCRIPTION, fill = Count)) +
  geom_tile(color = "white") +  # Use a white border for better visibility
  labs(title = paste0("Count of projects queries by user from ", from_date, " to ", to_date),
       x = "User (ISID)",
       y = "Project Description",
       fill = "Count") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
  #scale_fill_gradient(low = "lightblue", high = "darkblue")  # Color gradient for counts
  scale_fill_viridis_c()

ggsave(file="DEV_user_project_queries.png", plot=plt, width=16, height=12)
```

```r
df_summary <- df %>%
  group_by(ISID, PROJECTS_DESCRIPTION) %>%
  summarise(Count = n(), .groups = "drop") %>%
  mutate(ProjectShort = substr(PROJECTS_DESCRIPTION, 1, 6))  # Get first 6 letters

ggplot(df_summary, aes(x = ProjectShort, y = Count, fill = PROJECTS_DESCRIPTION)) +
  geom_bar(stat = "identity") +
  labs(title = "Count of queries of PROJECTS_DESCRIPTION by User (ISID)",
       x = "Project Description",
       y = "Count") +
  theme_minimal() +
  #theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
  theme( axis.text.x = element_blank()) +
  #scale_fill_discrete(name = "Project Description") +
  scale_fill_viridis_d() +
  facet_wrap(~ ISID, scales = "fixed") 
```

```r
df_summary_yearly <- df %>%
  mutate(Year = as.integer(substr(LAST_RUN, 1, 4))) %>%
  group_by(Year, PROJECTS_DESCRIPTION) %>%
  summarise(QueryCount = n(), .groups = "drop")

head(df_summary_yearly)
```

```r
plt <- ggplot(df_summary_yearly, aes(x = PROJECTS_DESCRIPTION, y = QueryCount, fill = PROJECTS_DESCRIPTION)) +
  geom_bar(stat = "identity", show.legend = FALSE) +  # Bar plot for each project description
  labs(title = "Total Queries by Project Description for Each Year",
       x = "Project Description",
       y = "Query Count") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1),
        panel.border = element_rect(color = "black", fill = NA),  # Add border to each facet
        panel.spacing = unit(1, "lines"),  # Increase space between panels
        panel.background = element_rect(fill = "white", color = "gray")) +
  scale_fill_viridis_d() +  # Optional: Use a color scale for better aesthetics
  facet_wrap(~ Year)  # Create a separate plot for each Year

ggsave(file="DEV_user_project_queries_by_year.png", plot=plt, width=16, height=12)
```

```r
query_counts <- df %>%
  group_by(ISID, PROJECTS_DESCRIPTION) %>%
  summarise(QueryCount = n(), .groups = "drop")
head(query_counts)
```

```r
ggplot(query_counts, aes(x = PROJECTS_DESCRIPTION, y = ISID, fill = QueryCount)) +
  geom_tile() +
  labs(title = "Heatmap of Queries by User and Project Description",
       x = "Project Description",
       y = "User (ISID)",
       fill = "Query Count") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

```r
ggplot(query_counts, aes(x = PROJECTS_DESCRIPTION, y = ISID)) +
  geom_point(aes(size = QueryCount, color = PROJECTS_DESCRIPTION), alpha = 0.6) +
  labs(title = "Bubble Plot of Queries by User and Project Description",
       x = "Project Description",
       y = "User (ISID)",
       color = "Project Description",
       size = "Query Count") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```
